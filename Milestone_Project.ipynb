```python
from alpha_vantage.timeseries import TimeSeries
import requests
import pandas as pd
import numpy as np
import bokeh
from bokeh.plotting import figure,output_file,show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool
import streamlit as st
```

```python
output_notebook()
```

```python
API_key = 'BGNKZDPGRQ7RCZ7L'
def get_data(symbol):
    ts = TimeSeries(key=API_key,output_format='pandas')
    data=ts.get_monthly_adjusted(symbol)[0]
    return data
```

```python


st.write("""
# TDI Milestone Example
An interactive chart of stock closing prices using Streamlit and Bokeh.
""")
st.sidebar.header('Select plot parameters:')

def user_input_features():
    ticker = st.sidebar.text_input("Ticker", 'JBLU')
    start_year = st.sidebar.text_input("Year", '2020')
    start_month = st.sidebar.text_input("Month", 'January')
    return ticker, start_year, start_month

symbol, start, end = user_input_features()

company_name = get_data(symbol.upper())

```

```python
st.header(f"{company_name} : Open, Close, Adjusted Close, High, Low Price" )

company=ticker
original_df=get_data(company)
df=original_df.reset_index()
df['date'] = pd.to_datetime(df['date'], format='%m/%d/%Y')

source = ColumnDataSource(df)
p = figure(x_axis_type="datetime")
p.line(x='date', y='1. open', line_width=1, source=source, color='red', legend=company+': Open')
p.line(x='date', y='4. close', line_width=1, source=source, color='orange', legend=company+': Close')
p.line(x='date', y='5. adjusted close', line_width=1, source=source, color='yellow', legend=company+': Adjusted Close')
p.line(x='date', y='2. high', line_width=1, source=source, color='green', legend=company+': High')
p.line(x='date', y='3. low', line_width=1, source=source, color='blue', legend=company+': Low')

p.title.text = 'Stock Price by Month/Year'
p.yaxis.axis_label = 'USD'
show(p)
```
